## 爬取完整內文
由於excerpt欄位中不包含完整內文，故抓出id後爬取
* URL: "https://www.dcard.tw/service/api/v2/posts/" + "ID"
* 只抓取部分需要欄位
  * id, title, content, createdAt, updatedAt, forumName, gender, likeCount, reactions, topics


### Write a function to crawl data

```python
def Crawl(ID):
    url_dcard = "https://www.dcard.tw/service/api/v2/posts/" + str(ID)
    requ = requests.get(url_dcard)
    rejs = requ.json()
    posts = dict()
    
    posts['ID'] = rejs['id']
    posts['title'] = rejs['title']
    posts['content'] = rejs['content']
    posts['createdAt'] = rejs['createdAt']
    posts['updatedAt'] = rejs['updatedAt'],
    posts['commentCount'] = rejs['commentCount']
    posts['forumName'] = rejs['forumName']
    posts['forumAlias']  = rejs['forumAlias']
    posts['gender'] = rejs['gender']
    posts['likeCount'] = rejs['likeCount']
    posts['topics'] = rejs['topics']

    return posts
```

```python
id_list = list(data['id'])
```

### 分次跑貼文
一次跑全部id很容易被擋,嘗試一次30篇

```python
# 容易被擋
posts_list = []
for id in id_list:
  posts_list.append(Crawl(id))
```

```python
# 嘗試每爬20篇休息25分鐘
from time import sleep
posts_list = []

for i in range(len(data)):
  print(i)
  if i > 0 and i%20 == 0:
    print("休25分鐘")
    sleep(1500)
  try:
    posts_list.append(Crawl(id_list[i]))
  except KeyError:
    pass
```

## 輸出成CSV檔

```python
df_post = pd.DataFrame(posts_list)
df_post
```

```python
df_post.to_csv('post_list.csv', encoding='utf_8_sig')
```
