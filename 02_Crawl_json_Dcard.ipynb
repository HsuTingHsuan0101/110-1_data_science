### Write a function to crawl data

```python
def Crawl(ID):
    url_dcard = "https://www.dcard.tw/service/api/v2/posts/" + str(ID)
    requ = requests.get(url_dcard)
    rejs = requ.json()
    posts = dict()
    
    posts['ID'] = rejs['id']
    posts['title'] = rejs['title']
    posts['content'] = rejs['content']
    posts['createdAt'] = rejs['createdAt']
    posts['updatedAt'] = rejs['updatedAt'],
    posts['commentCount'] = rejs['commentCount']
    posts['forumName'] = rejs['forumName']
    posts['forumAlias']  = rejs['forumAlias']
    posts['gender'] = rejs['gender']
    posts['likeCount'] = rejs['likeCount']
    posts['topics'] = rejs['topics']

    return posts
```

```python
id_list = list(data['id'])
```

```python
# å®¹æ˜“è¢«æ“‹
posts_list = []
for id in id_list:
  posts_list.append(Crawl(id))
```
